## Propuesta de arquitectura de red neuronal

Describe: 

+ las neuronas en la capa de entrada 
+ las capas intermedias – al menos dos – 
+ capa de salida
+ funciones de activación

La red neuronal propuesta tiene una estructura secuencial y consta de una capa de entrada, dos capas ocultas intermedias y una capa de salida. A continuación, se detalla cada componente de la arquitectura:
- Entrada: n neuronas, sin función de activación. En nuestro caso, n=27
- Primera capa intermedia: 64 neuronas, función de activación ReLU.
- Segunda capa intermedia: 64 neuronas, función de activación ReLU.
- Salida: 1 neurona, sin función de activación ( se debe a que necesitamos solo un dato para la regresión lineal).

Utilizamos además 10 epoch, función de pérdida el MSE y el optimizador RMSProp.

```python
# Código de la estructuración de la red
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Definir la arquitectura del modelo
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[X_train.shape[1]]),  # Capa de entrada
        layers.Dense(64, activation='relu'),  # Primera capa intermedia
        layers.Dense(1)  # Capa de salida
    ])
    
    # Configuración del optimizador
    optimizer = tf.keras.optimizers.RMSprop(0.001)

    # Compilación del modelo
    model.compile(loss='mse',                # Pérdida: Error cuadrático medio
                  optimizer=optimizer,       # Optimizador RMSProp
                  metrics=['mae', 'mse'])    # Métricas: Error absoluto medio y MSE
    return model

# Construir el modelo
model = build_model()
```

##  Ajuste de modelo de Regresión RNA

Mediante Python y utilizando al menos Keras sobre TensorFlow 2.0 (tensorflow.keras), entrena el modelo o modelos de red neuronal escogidos.

En el siguiente apartado, utilizaremos un pipeline que rellene los valores faltaantes con la media y que normalice los datos

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
# Preprocesamiento
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Rellena valores faltantes con la media
    ('scaler', StandardScaler())  # Escala los datos
])

# Aplicar transformaciones a todas las columnas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, X_train.columns)  # Todas las columnas
    ]
)

# Crear el pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Transformar los datos
X_train_transformed = pipeline.fit_transform(X_train)
X_test_transformed = pipeline.transform(X_test)
```

### Evalúe los resultados  para el conjunto de test.

```python
#Código de evaluación de la red propuesta (evaluación conjunto de test)
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  
  plt.legend()
  plt.show()


plot_history(history)
```

Se puede ver que los errores son muy grandes, lo cual ya indica que ambos modelos son poco eficaces al predecir los datos.
